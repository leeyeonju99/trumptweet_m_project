```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.drop(['mentions','hashtags','geo'],axis=1, inplace=True)
df.head()
```

# **트윗 반응 분석**

- 좋아요, 리트윗

```python
# 리트윗 top5
df.sort_values(by='retweets', ascending=False).head()
```

```python
df_fav=df.sort_values(by='favorites', ascending=False)
df_fav.head()
```

```python
#상위 1% 수 확인
df_fav_01 = int(len(df_fav) * 0.01)
df_fav_01
```

```python
df_fav_top=df_fav.head(231)
df_fav_top
```

```python
df_fav_top['content'] = df_fav_top['content'].str.replace('[^\w\s]+','')   #특수문자 제거거
df_fav_top
```

```python
tweets = df_fav_top['content'].to_list()
tweets
```

```python
# 앞에서 정제한 모든 트윗 텍스트를 하나의 문자열로 결합합니다. 
cleaned_text = ' '.join(tweets)
```

# **WordCloud 생성**

```python
# WordCloud 생성
from wordcloud import WordCloud, STOPWORDS 
wordcloud = WordCloud(width = 800, height = 400, 
                      background_color ='white', 
                      stopwords = set(STOPWORDS), 
                      max_words=100,
                      min_font_size = 10,
                      max_font_size = 100,
                     ).generate(cleaned_text) 
  
# WordCloud 출력                    
plt.figure(figsize = (8, 4), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout() 
plt.show()
```

# **감정 분석**

```python
from textblob import TextBlob
```

```python
df_sen=df_fav.head(231)
df_sen
```

```python
def get_sentiment(text):
    blob = TextBlob(text)
    sentiment = blob.sentiment.polarity
    if sentiment > 0:
        return 'positive'
    elif sentiment == 0:
        return 'neutral'
    else:
        return 'negative'

df_sen['sentiment'] = df_sen['content'].apply(get_sentiment)
df_sen
```

```python
sentiment_counts = df_sen['sentiment'].value_counts()
sentiment_counts.plot(kind='bar', color=['green', 'red', 'gray'])
```

```python
counts = df_sen['sentiment'].head(10).value_counts()
counts.plot(kind='bar', color=['red', 'gray', 'green'])
```
